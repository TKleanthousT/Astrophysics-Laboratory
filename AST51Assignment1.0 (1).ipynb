# Background


---
Welcome to our incredible tutorial on how to plot scatter plots with error bars in Python! Let's begin with a quick introduction to plotting.

**What is plotting?**
We plot data when we want to show the relationship between two variables. This can be anything: "How does the amount of sleep affect a student's grades?", "How does the mass of a star affect its luminosity?"; Essentially, "How does x affect y?"

**Okay, but then what is a scatter plot?**
Scatter plots are a type of graph that represents the relationship between two variables by displaying data points on a Cartesian coordinate system. They use "dots" to display the relationship between two sets of data. It is commonly used to observe trends, correlations, and distributions in data!

**What about error bars?**
Error bars are the graphical representation of the variability within the data we are plotting. It is often used to indicate uncertainty, standard deviation, or confidence intervals. It contextualizes the data, letting us know how reliable our measurements are by visually displaying the possible range of error.



**Hmm...so when do we use them?**
Scatter plots are especially useful when you have a lot of individual data points, because then we can visually analyze the relationship between the two variables! Let's give an example:

## Scatter Plot

How do we plot a scatter plot? Even further, how do we plot a scatter plot with errorbars? We have two options:


The `scatter()` function creates a simple scatter plot **without** error bars. It takes the following arguments:

- **`x`**: The x-coordinates of the data points.
- **`y`**: The y-coordinates of the data points.
- **`marker`**: Sets the shape of the data points (e.g., `'o'`, `'s'`, `'^'`, etc.).
- **`color`**: Specifies the color of the data points.
- **`s`**: Sets the size of the data points.
- **`label`**: A label for the dataset (used in the legend).

The `errorbar()` function creates a scatter plot with error bars, allowing you to visualize uncertainties in your data. It accepts the following arguments:

- **`x`**: The x-coordinates of the data points.
- **`y`**: The y-coordinates of the data points.
- **`xerr`**: The horizontal error margin (scalar or array-like).
- **`yerr`**: The vertical error margin (scalar or array-like).
- **`fmt`**: A format string that defines the style and shape of the data points (e.g., `'o'`, `'s'`, etc.).
- **`color`**: Specifies the color of the data points.
- **`ecolor`**: Specifies the color of the error bars.
- **`alpha`**: Sets the transparency level of the data points (0.0 is fully transparent, 1.0 is fully opaque).
- **`label`**: A label for the dataset (used in the legend).
- **`capsize`**: Sets the size (in points) of the horizontal lines (caps) at the top and bottom of the error bars.

Considering the error that concerns us in Stanley's case is the error surrounding his weight measurements, we will be using vertical errorbars rather than horizontal errorbars. Always remember to plot the appropriate errorbars for your data. Are you concerned with the uncertainty in the x or in the y? Maybe both?

```python
# Create scatter plot with vertical error bars
plt.errorbar(months, MeasuredWeight, yerr=MeasurementError, fmt='o', color="blue", alpha=0.6, label="Measured Weight", capsize=5)
```

Great! But let's not forget that every graph needs axis labels, a title, and a legend. However, when we put it all together a couple of cells down from here, everything will run smoothly! Let's even add a grid for readability:

```python
# Labels and title
plt.xlabel("Months", fontsize=16)
plt.ylabel("Weight (kg)", fontsize=16)
plt.title("Stanley's Weight Over Twelve Months", fontsize=18)
plt.legend(fontsize=12)  # Show legend
plt.grid(True)  # Add grid for better readability
```

Perfect! Finally, in order to actually show the plot, we must run the `show()` function.

```python
# Show plot
plt.show()
```

Now, put it all together:

```python
# Define the figure size
plt.figure(figsize=(8, 5))

# Create scatter plot with error bars
plt.errorbar(months, MeasuredWeight, yerr=MeasurementError, fmt='o', color="blue", alpha=0.6, label="Measured Weight", capsize=5)

# Labels and title
plt.xlabel("Months", fontsize=16)
plt.ylabel("Weight (kg)", fontsize=16)
plt.title("Stanley's Weight Over Twelve Months", fontsize=18)
plt.legend(fontsize=12)  # Show legend
plt.grid(True)  # Add grid for better readability

# Show plot
plt.show()
```

## Line of Best Fit
Interesting! Let's take a look at the overall trend by adding a line of best fit. To do so we are going to use the `polyfit()` function from numpy. It will do, in our case, a linear regression of the data to find the line of best fit. What `polyfit()` does is that it returns the coefficients for a polynomial of a certain degree that is a best fit for our data. For a linear regression, we would need two coefficients, which we have defined below as "m" and "b" for y=mx+b. On the other hand, if we were to create a quadratic regression, we would need 3 coefficients instead of two, since your equation is now in quadratic form! Using a higher order line of best fit can sometimes be more accurate because it captures characteristics that lower order lines cannot, but keep in mind that it also requires more coefficients!

```python
# Define the figure size
plt.figure(figsize=(8, 5))

# Create scatter plot with error bars
plt.errorbar(months, MeasuredWeight, yerr=MeasurementError, fmt='o', color="blue", alpha=0.6, label="Measured Weight", capsize=5)

# Add line of best fit, use polyfit() function in numpy
m, b = np.polyfit(months, MeasuredWeight, 1)  # Linear regression (1st-degree polynomial)
plt.plot(months, m * months + b, color="red", linestyle="--", label="Best Fit Line")

# Labels and title
plt.xlabel("Months", fontsize=16)
plt.ylabel("Weight (kg)", fontsize=16)
plt.title("Stanley's Weight Over Twelve Months", fontsize=18)
plt.legend(fontsize=12)  # Show legend
plt.grid(True)  # Add grid for better readability

# Show plot
plt.show()
```

# Good Plotting Practices


---

Now that we know how to make a simple scatter plot thanks to English David, let's talk asthetics! When working with data visualization in Python, it is essential to create graphs that are both informative and visually appealing.

Somethings to keep in mind when trying to hold good plotting practices:
* Choose clear and meaningful labels
* Use a logical scale
* Select the right marker style and size
* Add error bars when needed
* Use color wisely
* Add a legend when necessary
* Keep gridlines and background simple

Let's be cognizant of this as we model the perfect, noiseless motion of a ball bounce:

$\text{height} = -0.5t^2 + 5t$

## Nonphysical Plot

In this case, our x-array describes time and our y-array describes the motion. Let's use this information to create a good plot.

```python
# Define xs, aka time array
# We are never at negative time so we begin the array at 0 rather than a negative number, 0 being the moment the ball leaves the ground
time = np.linspace(0, 20, 25)

# Define ys, aka height array
height = (-0.5 * time**2) + (5 * time)

# This is an ideal scenario so no error is needed

# Define the figure size
plt.figure(figsize=(8, 5))

# Plot data
plt.scatter(time, height, marker='o', s = 30, label='Ball', color='b')

# Add labels and title
plt.xlabel('Time (seconds)', fontsize=16)
plt.ylabel('Height (feet)', fontsize=16)
plt.title('Perfect, Noiseless Motion of a Ball Bounce', fontsize=18)

# Add grid and legend
plt.grid(True, linestyle='-', alpha=0.6)
plt.legend(fontsize=12)

# Show the plot
plt.show()
```

## Physical Plot
Notice that what the graph is depicting is not physical because we see a negative height once we pass 10 seconds. Let's zoom in to show only the physical result of the ball bouncing. You can do this by modifying the arguments of the `xlim()` and `ylim()` as they set the limits for what the x-axis and y-axis will display.

```python
# Define xs, aka time array
# We are never at negative time so we begin the array at 0 rather than a negative number, 0 being the moment the ball leaves the ground
# Let's add more data points so we can properly see the trend of the data in this region
time = np.linspace(0, 20, 50)

# Define ys, aka height array
height = (-0.5 * time**2) + (5 * time)

# This is an ideal scenario so no error is needed

# Define the figure size
plt.figure(figsize=(8, 5))

# Plot data
plt.scatter(time, height, marker='o', s = 30, label='Ball', color='b')

# Add labels and title
plt.xlabel('Time (seconds)', fontsize=16)
plt.ylabel('Height (feet)', fontsize=16)
plt.title('Perfect, Noiseless Motion of a Ball Bounce', fontsize=18)

# Use xlim and ylim to zoom into the physical part of the graph
# You can contract or extend these as you see fit
plt.xlim(np.min(time) - 1, 11)
plt.ylim(0, np.max(height) + 1)

# Add grid and legend
plt.grid(True, linestyle='-', alpha=0.6)
plt.legend(fontsize=12)

# Show the plot
plt.show()
```

Now look at that! We can fully see our quadratic function in action! All the labels are a good size, the data points have a clear marker shape, and we removed the error bars because they were not relevant to our perfect, noiseless motion.
